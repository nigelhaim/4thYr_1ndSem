# General path

```python
# Install the required libraries
# %pip install matplotlib

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.path as path
```

```python
vertices = [
    (50,50),
    (50,200),
    (150, 500), (250, 200),
    (350, 100), (150, 150), (100, 100),
    (50, 50)
]

codes = [
    path.Path.MOVETO,
    path.Path.LINETO,
    path.Path.CURVE3, path.Path.CURVE3,
    path.Path.CURVE4, path.Path.CURVE4, path.Path.CURVE4,
    path.Path.CLOSEPOLY, 
]
```

```python
# Creating the path 
p = path.Path(vertices, codes)

# Create the patch to display the path 
patch = patches.PathPatch(p, facecolor='yellow', edgecolor = 'blue', lw=2)

#plot the image 
fig, ax = plt.subplots()
ax.add_patch(patch)
ax.set_xlim(0, 500)
ax.set_ylim(0, 500)
plt.show()
```

```python
# Drawinga car in python with gen path 
vertices_car = [
    (60, 120), 
    (80, 120),
    (90, 140), (100, 120),
    (160, 120),
    (170, 140), (180,120),
    (200,120),
    (195, 100), (200, 80), (160,80), (100, 80),
    (100, 80), 
    (60, 100), 
    (60, 120)
]

codes_car = [
    path.Path.MOVETO, 
    path.Path.LINETO,
    path.Path.CURVE3, path.Path.CURVE3,
    path.Path.LINETO,
    path.Path.CURVE3, path.Path.CURVE3,
    path.Path.LINETO,
    path.Path.CURVE4, path.Path.CURVE4, path.Path.CURVE4,
    path.Path.LINETO, path.Path.LINETO, path.Path.LINETO,
    path.Path.CLOSEPOLY
]
```
