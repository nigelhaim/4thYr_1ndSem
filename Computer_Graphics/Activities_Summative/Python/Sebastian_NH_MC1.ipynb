## Draw straight line

```python
A = (50, 475)
B = (500, 125)
C = (300, 300)
D = (450, 500)
E = (200, 375)

draw.line((A, B), fill = lines_color, width = line_width)
draw.line((B, C), fill = lines_color, width = line_width)
draw.line((C, D), fill = lines_color, width = line_width)
draw.line((D, E), fill = lines_color, width = line_width)
draw.line((E, A), fill = lines_color, width = line_width)

draw.text(B, 'Line1', fill = String_color, font = font)
draw.text(C, 'Line2', fill = String_color, font = font)
draw.text(D, 'Line3', fill = String_color, font = font)
draw.text(E, 'Line4', fill = String_color, font = font)
draw.text(A, 'Line5', fill = String_color, font = font)
```

## Draw Quad Curve

```python

p0 = (175, 500)
c1 = (200, 300)
p2 = (400, 550)
t = 0.5

def quad_curve(t, p0, p1, p2):
    x = (1-t)**2 * p0[0] + 2 * (1-t) * t * c1[0] + t**2 * p2[0]
    y = (1-t)**2 * p0[1] + 2 * (1-t) * t * c1[1] + t**2 * p2[1]
    return x, y
points = [quad_curve(t, p0, c1, p2) for t in [i/100 for i in range(101)]]
draw.line(points, fill = QuadraticCurve_color, width = line_width)
draw.text(p0, 'Quadratic Curve', fill = String_color, font = font)
```

## Draw Cubic Curve

```python
p0 = (375,300)
p1 = (450,550)
p2 = (550,450)
p3 = (575,350)

def cubic_bezier(t, p0, p1, p2, p3):
  x = (1-t)**3 * p0[0] + 3 * (1-t)**2 * t * p1[0] + 3 * (1-t) * t**2 * p2[0] + t**3 * p3[0]
  y = (1-t)**3 * p0[1] + 3 * (1-t)**2 * t * p1[1] + 3 * (1-t) * t**2 * p2[1] + t**3 * p3[1]
  return x, y
points = [cubic_bezier(t, p0, p1, p2, p3) for t in [i/100 for i in range(101)]]
draw.line(points, fill = CubicCurve_color, width = 5)
draw.text(p0, 'Cubic Curve', fill = String_color, font = font)
```
