#### Quad curve

```python
p0 = (50, 300)
p1 = (250, 150)
p2 = (350, 200)
t = 0.5

def quad_curve(t, p0, p1, p2):
    x = (1-t)**2 * p0[0] + 2 * (1-t) * t * p1[0] + t**2 * p2[0]
    y = (1-t)**2 * p0[1] + 2 * (1-t) * t * p1[1] + t**2 * p2[1]
    return x, y
points = [quad_curve(t, p0, p1, p2) for t in [i/100 for i in range(101)]]
quad = draw.line(points, fill = blue, width = line_width)
```

#### Cubic Curve

```python
p0 = (50, 300)
p1 = (250, 375)
p2 = (350, 150)
p3 = (450, 200)

def cubic_bezier(t, p0, p1, p2, p3):
  x = (1-t)**3 * p0[0] + 3 * (1-t)**2 * t * p1[0] + 3 * (1-t) * t**2 * p2[0] + t**3 * p3[0]
  y = (1-t)**3 * p0[1] + 3 * (1-t)**2 * t * p1[1] + 3 * (1-t) * t**2 * p2[1] + t**3 * p3[1]
  return x, y

points = [cubic_bezier(t, p0, p1, p2, p3) for t in [i/100 for i in range(101)]]
bicubic = draw.line(points, fill = pink, width = line_width)
```
